# Pendahuluan

*Video game* telah menjadi hiburan bagi kita sejak era 70-an ditandai dengan dijualnya konsol *game* Atari secara masal. Sejak saat itu *video game* menjadi bagian dari industri. Semakin banyak *game console* terjual ke seluruh dunia. Semakin banyak pula *game developer* yang membuat dan menjual *game* di seluruh dunia.

Berbaga *genre* untuk berbagai *game console* bermunculan karena permintaan pasar. Oleh sebab itu, *game-game* terbagi ke dalam segmentasi *rating* ESRB demi kenyamanan pengguna yang terdiri dari berbagai umur.

Pada analisis ini, pasar-pasar tersebut terbagi menjadi pasar Amerika Utara, Eropa, Jepang dan negara lain. Analisis ini mengidentifikasi pola-pola yang menentukan apakah suatu *game* berhasi atau tidak, mempelajari trend penjualan *game*, berapa lama suatu *game console* dapat bertahan sebelum punah,  distribusi penjualan *game* berdasarkan *genre* pada tiap pasar, *genre* dan *game console* favorit pada tiap pasar, pengaruh penilaian *game* terhadap penjualannya, pengaruh *rating* ESRB terhadap penjualan *game*, serta menjawab hipotesis rata-rata penjualan *game* pada dua *game console* adalah sama dan hipotesis penjualan *game* pada dua *genre* adalah sama.

### Penjualan Game pada Pasar Amerika Utara Berdasarkan Rating ESRB

Untuk memudahkan memahami *rating* ESRB, maka berikut penjelasannya.

#### Kategori *Rating* ESRB
- EC (*Early Childhood*): konten yang cocok untuk anak umur 3 tahun ke atas. Tidak mengandung apapun yang tidak pantas menurut orang tua bagi anak umur 3 tahun ke atas.
- K-A (*Kids to Adult*): konten yang cocok untuk semua umur. *Rating* ini muncul pada awal terbentuknya ESRB pada tahun 1994, kemudian diganti menjadi E (*Everyone*) pada tahun 1998.
- E (*Everyone*): konten yang cocok untuk semua umur. Dapat mengandung kartun, fantasi, *mild violence* (kekerasan realistis yang tidak berujung pada pendarahan atau kematian), dan sesekali penggunaan *mild language* (bahasa yang mengandung kata-kata kasar, seksual, kekerasan, dan penggunaan alkohol dan obat-obatan).
- E10+ (*Everyone* 10+): konten yang cocok untuk umur 10 tahun ke atas. Dapat mengandung lebih banyak kartun, fantasi, *mild violence*, *mild language* atau *suggestive themes* dengan kadar rendah.
- T (*Teen*): konten yang cocok untuk umur 13 tahun ke atas. Dapat mengandung kekerasan, *suggestive themes*, *crude humor*, konten berdarah dengan kadar rendah, atau sesekali penggunaan *strong language*.
- M (*Mature* 17+): konten yang cocok untuk umur 17 tahun ke atas. Dapat mengandung *intense violence*, tumpahan atau cipratan darah karena kekerasan, konten seksual, atau *strong language*.
- AO (*Adults Only*): konten yang coock untuk umur 18 tahun ke atas. Dapat mengandung adegan berlanjut dan panjang kekerasan, gambaran *sexual content* atau perjudian dengan mata uang asli.
- RP (*Rating Pending*): belum memiliki keputusan final dalam *rating* ESRB. Simbol ini hanya muncul dalam materi pengiklanan dan promosi sebelum *game* rilis, dan harus sudah tergantikan ketika *game* rilis.

#### Definisi Istilah pada *Rating* ESRB
- *mild violence*: kekerasan realistis yang tidak mengakibatkan pendarahan atau kematian.
- *mild language*: bahasa yang mengandung kata-kata kasar, seksual, kekerasan, dan penggunaan alkohol dan obat-obatan.
- *strong language*: penggunaan bahasa yang mengandung kata-kata kasar dalam frekuensi yang sering.
- *suggestive themes*: referensi atau material seksual.
- *crude humor*: gambaran atau dialog dengan perilaku vulgar, termasuk humor "kamar mandi".
- *intese violence*: gambaran realistis benturan fisik. Dapat mengandung darah secara ekstrim dan realistis, cipratan darah karena kekerasan, senjata dan gambaran cidera atau kematian manusia.

Berdasarkan penjelasan di atas, *rating* Everyone (E) adalah pengganti untuk *rating* Kids to Adult (K-A), maka semua *game* dengan *rating* Kids to Adult (K-A) akan diganti menjadi Everyone (E).

## Menguji Hipotesis

### Rata-Rata Rating Pengguna antara *Platform* Xbox One dan PC adalah Sama

- H0 (*Null Hypothesis*)= rata-rata penilaian pengguna antara *platform* Xbox One dan PC adalah sama.
- H1 (*Alternative Hypothesis*)= rata-rata penilaian pengguna antara *platform* Xbox One dan PC adalah beda.

```python
# menyiapkan dataset yang hanya berisi platform Xbox One dan tanpa penilaian pengguna yang kosong.
df_xbox_one=df_new_2[df_new_2['platform']=='XOne']
df_xbox_one=df_xbox_one[df_xbox_one['user_score'].notnull()].reset_index(drop=True)
df_xbox_one
```

```python
# menyiapkan dataset yang hanya berisi platform PC dan tanpa penilaian pengguna yang kosong.
df_pc=df_new_2[df_new_2['platform']=='PC']
df_pc=df_pc[df_pc['user_score'].notnull()].reset_index(drop=True)
df_pc
```

```python
# menjalankan Levene Test
alpha=0.05
result=st.levene(df_xbox_one['user_score'], df_pc['user_score'])
print('nilai p value:', result.pvalue)
print('nilai alpha:', alpha)

if result.pvalue<alpha:
    print('Kita menolak H0 (Null Hypothesis)')
else:
    print('Kita tidak bisa menolak H0 (Null Hypothesis)')
```

#### Kesimpulan
Bisa kita lihat bahwa kita tidak bisa menolak H0, maka bisa disimpulkan rata-rata *rating* pengguna antara *platform* Xbox One dan PC adalah sama.
